```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Iris.csv')
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    x='SepalLengthCm',
    y='PetalLengthCm',
    hue='Species',
    data=df
)

plt.title("Sepal Length vs Petal Length")
plt.show()
```

The scatter plot shows a positive relationship between sepal length and petal length. Different species form distinct clusters, making species classification easier.

```python
plt.figure(figsize=(8,6))

plt.hist(df['SepalLengthCm'], bins=10)

plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")

plt.show()
```

The histogram shows the distribution of sepal length values. Most observations are concentrated around the center, indicating a roughly normal distribution.

```python
plt.figure(figsize=(8,6))

sns.boxplot(
    x='Species',
    y='PetalLengthCm',
    data=df
)

plt.title("Petal Length by Species")

plt.show()
```

The box plot compares petal lengths across species. It highlights differences in median values and helps identify potential outliers.

```python
print(df.isnull().sum())
```

No missing values were found in the dataset.

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(8,6))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Matrix")

plt.show()
```

Petal length and petal width have a strong positive correlation, indicating they increase together.
